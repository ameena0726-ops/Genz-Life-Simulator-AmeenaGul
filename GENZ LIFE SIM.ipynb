```python
# ğŸ’¬ Assignment 1 â€“ Personal AI Chatbot (AIC-202)
# Project Title: GenZ Life Simulator
# Student Name: Ameena Gul
# Student ID: 01-136242-051

import random

# ---------- 1. Introduction ----------
def chatbot_intro():
    print("Hey bestie ğŸ‘‹ Welcome to the GenZ Life Simulator!")
    user_name = input("Enter your name: ")
    user_age = input("Enter your age: ")
    print(f"Nice to meet you, {user_name}! At {user_age}, youâ€™re absolutely thriving ğŸ˜âœ¨")
    print("Iâ€™m your virtual buddy â€” letâ€™s live your best GenZ life together ğŸ’–")
    return user_name

# ---------- 2. Response Section ----------
responses = {
    "hi": "Hey bestie! ğŸ’…",
    "hello": "Hii ğŸ‘‹ You look ready to slay today!",
    "how are you": "Vibing and surviving ğŸ˜Œ What about you?",
    "how's your day going?":"Quite good . Wbu?",
    "bye" : "Aww okay bye bestie ğŸ’– Stay iconic!",
    "thanks": "Anytime bestie âœ¨",

}

def get_response(message):
    message = message.lower()
    if message in responses:
        return responses[message]
    elif "play" in message:
        return "Do you wanna start your GenZ Life Simulator? (yes/no)"
    else:
        return random.choice([
            "idk what that means but it sounds cool ğŸ˜‚",
            "lol thatâ€™s random but I love the energy ğŸ˜…",
            "say that again but with âœ¨vibesâœ¨"
        ])

# ---------- 3. Player Information ----------
player_stats = {"energy": 100, "points": 0, "mood": "ğŸ˜Š"}

# ---------- 4. Unique Feature: Cook System ğŸ³ ----------
def cooking_feature():
    print("\nğŸ‘©â€ğŸ³ Welcome to your virtual kitchen!")
    print("Choose what you want to cook today:")
    print("1ï¸âƒ£ Pasta ğŸ  2ï¸âƒ£ Pancakes ğŸ¥  3ï¸âƒ£ Mystery Dish ğŸ§ª")

    choice = input("Enter your choice (1â€“3): ")

    if choice == "1":
        print("You made delicious pasta ğŸ â€” chef vibes only!")
        player_stats["energy"] += 15
        player_stats["points"] += 10
        player_stats["mood"] = "Satisfied ğŸ˜‹"

    elif choice == "2":
        print("You flipped perfect pancakes ğŸ¥ like a pro!")
        player_stats["energy"] += 20
        player_stats["points"] += 5
        player_stats["mood"] = "Sweet and cozy ğŸ’•"

    elif choice == "3":
        print("You tried making a mystery dish ğŸ§ª...")
        result = random.choice(["ğŸ’¥ Disaster!", "ğŸŒŸ Masterpiece!", "ğŸ”¥ Burnt mess!"])
        print(result)
        if "Masterpiece" in result:
            player_stats["points"] += 25
            player_stats["mood"] = "Proud ğŸ‘‘"
        elif "Burnt" in result:
            player_stats["energy"] -= 10
            player_stats["mood"] = "Okay-ish ğŸ˜…"
        else:
            player_stats["energy"] -= 20
            player_stats["mood"] = "Stressed ğŸ˜­"

    else:
        print("Thatâ€™s not on the menu bestie ğŸ˜…")

    # Keep stats valid
    player_stats["energy"] = max(0, min(100, player_stats["energy"]))
    print(f"âš¡ Energy: {player_stats['energy']} | ğŸ¯ Points: {player_stats['points']} | ğŸ˜„ Mood: {player_stats['mood']}")

# ---------- 5. Activities Section ----------
def perform_action(action):
    if action == "read":
        player_stats["energy"] -= 10
        player_stats["points"] += 10
        player_stats["mood"] = "Smart ğŸ“–"
        print("You read an article and learned something new ğŸ§ ")

    elif action == "sleep":
        player_stats["energy"] = 100
        player_stats["mood"] = "Rested ğŸ˜´"
        print("You took a nap and now feel recharged âš¡")

    elif action == "hangout":
        player_stats["energy"] -= 15
        player_stats["points"] += 5
        player_stats["mood"] = "Social ğŸ§‹"
        print("You hung out with your friends and had a great time ğŸ‰")

    elif action == "cook":
        cooking_feature()

    else:
        print("Hmm, thatâ€™s not something we can do right now ğŸ˜…")

    # Display stats
    player_stats["energy"] = max(0, min(100, player_stats["energy"]))
    print(f"âš¡ Energy: {player_stats['energy']} | ğŸ¯ Points: {player_stats['points']} | ğŸ˜„ Mood: {player_stats['mood']}")

# ---------- 6. Mini-Game Loop ----------
def start_simulator():
    print("\nğŸ® Game Started! Time to live your digital GenZ life ğŸ’–")
    while True:
        action = input("Choose an action (read/sleep/hangout/cook/quit): ").lower()
        if action == "quit":
            print("\nâœ¨ Simulation Ended âœ¨")
            print(f"Final Stats âœ âš¡ {player_stats['energy']} | ğŸ¯ {player_stats['points']} | ğŸ˜„ Mood: {player_stats['mood']}")
            print("You totally slayed this virtual life ğŸ’…")
            break
        perform_action(action)

# ---------- 7. Chatbot Interaction ----------
def start_chatbot():
    chatbot_intro()
    while True:
        message = input("\nYou: ")
        reply = get_response(message)
        print("Bot:", reply)

        if "wanna start" in reply:
            answer = input("You: ").lower()
            if answer == "yes":
                start_simulator()
            else:
                print("Alright bestie, maybe next time ğŸ’¬")

        elif message.lower() == "bye":
            print("Bye bestie ğŸ’– Have a great day IRL âœ¨")
            break

# ---------- 8. Run the Chatbot ----------
start_chatbot()
```
